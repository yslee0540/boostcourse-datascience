## Q2

```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
```

```python
df = df[df["Year"] >= 2011]
df
```

```python
# groupby
df.groupby(["Year", "Country"])["Life_Expectancy"].mean().unstack()
```

```python
# pivot_table
pd.pivot_table(df, index="Year", columns="Country", values="Life_Expectancy")
```
